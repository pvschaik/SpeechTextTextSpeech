```python
file_path = "nl_InternetAdvertising_en.txt"
# file_path = "nl_InternetAdvertising.txt"
file_path = Path(file_path)
file_path = path_data / file_path
text = file_path.read_text(encoding="utf-8")
print(text)
```

```python
# count sentences in source text, sumy
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer

parser = PlaintextParser.from_string(text, Tokenizer("english"))
# parser = PlaintextParser.from_string(text, Tokenizer("dutch"))
total_sentences = len(parser.document.sentences)
print(total_sentences)
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

def summarize(pText, pSentence_rate = 10, pLanguage = "english"):
    summary_sentence_count = int(round(total_sentences * (pSentence_rate/100),0))
    # parser = PlaintextParser.from_string(text, Tokenizer("english"))
    parser = PlaintextParser.from_string(pText, Tokenizer(pLanguage))
    summarizer = LsaSummarizer()
    summary = summarizer(parser.document, summary_sentence_count)
    summary_parsed =  " ".join(str(sentence) for sentence in summary)
    parser = PlaintextParser.from_string(summary_parsed, Tokenizer("english"))
    total_text_summarised = len(parser.document.sentences)
    print(total_text_summarised)
    
    return summary_parsed

print("summarize function defined")
```
